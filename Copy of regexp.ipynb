# Регулярные выражения

`или` в регулярных выражениях: напишите регулярку, которая находит всех котиков.

```python
import re
```

```python
expample = 'Я пришла на каток, а там был котик! Я люблю котиков, а этот был очень милый. Уруру, котеек!'
```

```python
re.findall ("кот.+?к(?:w+)?", expample)
```
